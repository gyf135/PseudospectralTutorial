An easy stroll through pseudospectral methods.




First load the package for plotting.

```julia
using PyPlot
```

Create the physical grid $x$. For a domain of length $L_x$ with $n_x$ grid points the grid spacing is $\delta=L_x/n_x$. Our grid points then are: $x = 0,\delta, 2\delta,\dots,L_x-\delta$. Notice, that the final grid point *does not* correspond to $x=L_x$ but to $x=L_x-\delta$. 

```julia
nx = 64
Lx = 2.0*pi

dx = Lx/nx
 x = 0:dx:Lx-dx

k  = 2.0*pi/Lx * [0:nx/2; -nx/2+1:-1];
```

Let's take now a known function $f(x)$ defined on the grid and see how we can calculate its derivative numerically. We shall first take a function for which we have a closed form for its derivative so that we can check our numerical results. (Although this step may seem obvious still many people oversee it...)

We have to make sure that our function is periodic with period $L_x$. Since we took $L_x=2\pi$ then all sines and cosines with integer wavenumbers satisfy the periodicity.

```julia
       f = sin.(2*x) + 0.5*cos.(4*x)   # a test function f(x)
 dfdx_an = 2*cos.(2*x) - 2*sin.(4*x)   # the analytic expression for df/dx

      fh = fft(f)                      # the FFT of f(x)
dfdx_num = real(ifft(im*k.*fh));       # the numerical approximation for df/df
```

Let's plot first the fft of $f(x)$. Do we see power at the wavenumbers we expected?

```julia
figure(1, figsize=(8, 3) )
plot(k, abs.(fh), marker=6, linestyle="dashed")  # plots the abs value of fft of f(x)
xlim(-14, 14)
xlabel(L"k_x")
ylabel(L"|\hat{f}_{k_x}|");
```

Now let's compare the numerical approximation of $\textrm{d}f/\textrm{d}x$ with the analytical one.

```julia
fig, axs = subplots(nrows=2, ncols=1, figsize=(8, 6) )
axs[1][:plot](x, dfdx_an, label="analytical")
axs[1][:plot](x, dfdx_num, linestyle="dashed", label="numerical")
axs[1][:set_xlabel](L"x")
axs[1][:set_ylabel](L"\textrm{d}f(x)/\textrm{d}x")
axs[1][:legend](loc="upper right", fancybox="true")
axs[1][:set_xlim](0, Lx-dx)
axs[2][:plot](x, abs.(dfdx_an - dfdx_num))
axs[2][:set_xlabel](L"x")
axs[2][:set_ylabel]("absolute error")
axs[2][:set_xlim](0, Lx-dx);
```

Oftentime people get the impression that pseudospectral methods only work for sines and cosine functions. That's not true. Let's try to compute the derivative of a gaussian. 

First we define a function, named gsn, that gives us the gaussian centered at $x_0$ with variance $\sigma$.

```julia
sig = 0.5
x0  = 2
gsn(x0, x) = 1./(sig*sqrt.(2.0*pi)).*exp.(-(x-x0).^2.0/(2.0*sig^2.0));
```

We then construct $f(x)$ by calling $\texttt{gsn}(x_0,x)$. We calculate its derivative using fft and compare it to the analytical solution.

```julia
       f = gsn(x0,x)
 dfdx_an = -(x-x0)/sig^2.*gsn(x0, x)

      fh = fft(f)

dfdx_num = ifft(im*k.*fh)

fig, axs = subplots(nrows=3, ncols=1, sharex=true, figsize=(8, 10))
axs[1][:plot](x, f)
axs[1][:set_ylabel](L"f(x)")
axs[2][:plot](x, dfdx_an, label="analytical")
axs[2][:plot](x, dfdx_num, linestyle="dashed", label="numerical")
axs[2][:set_xlabel](L"x")
axs[2][:set_ylabel](L"\textrm{d}f(x)/\textrm{d}x")
axs[2][:legend](loc="upper right", fancybox="true")
axs[2][:set_xlim](0, Lx-dx)
axs[3][:plot](x, abs.(dfdx_an - dfdx_num))
axs[3][:set_xlabel](L"x")
axs[3][:set_ylabel]("absolute error")
axs[3][:set_xlim](0, Lx-dx);
```

Why we so big error? The gaussian vanishes at the boundaries so it is periodic. 

Because we chose $x_0\ne L_x/2$ we have introduced an assymetry that violates the periodicity condition. To see that, we print the values of $f$ at the first and last grid points.

```julia
println(gsn(x0,0))
println(gsn(x0,2.0*pi))
```

We can make a periodic continuation of our gaussian so that *(i)* it is essentially the same as our initial gaussian whose derivative we want to approximate but still *(ii)* satisfy the periodicity condition. Thus, we construct

$$f_2(x) = \texttt{gsn}(x_0, x) + \texttt{gsn}(x_0+L_x, x) + \texttt{gsn}(x_0-L_x, x)\ .$$

We then proceed computing its fft and its derivative. In the first plot, we compare $f(x)$ and $f_2(x)$.

```julia
  f2 = gsn(x0, x) + gsn(x0-Lx, x) + gsn(x0+Lx, x)
df2dx = -(x-x0)/sig^2.*gsn(x0, x) -(x-x0-Lx)/sig^2.*gsn(x0+Lx, x) -(x-x0+Lx)/sig^2.*gsn(x0-Lx, x)

  fh = fft(f2)

dfdx_num = ifft(im*k.*fh)

fig, axs = subplots(nrows=3, ncols=1, sharex=true, figsize=(8, 10))
axs[1][:plot](x, f2, label=L"$f_2(x)$, periodic")
axs[1][:plot](x, f, linestyle="dashed", label=L"$f(x)$, the naive gaussina")
axs[1][:set_ylabel](L"f(x)")
axs[1][:legend](loc="upper right", fancybox="true")
axs[2][:plot](x, dfdx, label="analytical")
axs[2][:plot](x, dfdx_num, linestyle="dashed", label="numerical")
axs[2][:set_xlabel](L"x")
axs[2][:set_ylabel](L"\textrm{d}f(x)/\textrm{d}x")
axs[2][:legend](loc="upper right", fancybox="true")
axs[2][:set_xlim](0, Lx-dx)
axs[3][:plot](x, abs.(dfdx-dfdx_num))
axs[3][:set_xlabel](L"x")
axs[3][:set_ylabel]("absolute error")
axs[3][:set_xlim](0, Lx-dx);
```
